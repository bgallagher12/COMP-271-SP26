# Stacks and Queues

This notebook covers classes [`LinkedList`](./linked_list.py), [`EnhancedLinkedList`](./enhanced_linked_list.py), [`Stack`](./stack.py), and [`Queue`](./queue.py). All these classes require [`Node`](./node.py) objects.

We are implementing stack and queues using a linked list as the underlying data structure. The abstract data types are shown below.

```text
+--------------------------+            +-----------------------+
| Queue ADT                |            | Stack ADT             |
+--------------------------+            +-----------------------+
| enqueue(Node) -> None    |            | push(Node) -> None    |
| dequeue() -> Node | None |            | pop() -> Node | None  |
| peek() -> Node | None    |            | peek() -> Node | None |
| is_empty() -> bool       |            | is_empty() -> bool    |
| size() -> int            |            | size() -> int         |
+--------------------------+            +-----------------------+
```

## Contract for the `Queue` abstract data structure

* `enqueue`: add a new node to the end of the underlying data structure.

* `dequeue`: remove and return the node from the front of the underlying data structure.

* `peek`: return a copy of the node from the front of the underlying data structure.

* `is_empty`: return `True` when there are no nodes in the underlying data structure and `False` otherwise.

* `size`: return the number of nodes in the underlying data structure.

## Contract for the `Stack` abstract data structure

* `push`: add a new node to the front of the underlying data structure.

* `pop`: remove and return the node from the front of the underlying data structure.

* `peek`: return a copy of the node from the front of the underlying data structure.

* `is_empty`: return `True` when there are no nodes in the underlying data structure and `False` otherwise.

* `size`: return the number of nodes in the underlying data structure.

## Analysis of the contracts

There are immediate similarities between the two contracts. For example, methods `is_empty` and `size` appear identical. 

Method `peek` appears so too, but only because we chose the front of the underlying data structure for adding and removing data in a stack. If we have selected the back of the underlying data structure to add and remove data in the stack, the specifications would have been different:

* `Queue.peek`: return a copy of the node from the **front** of the underlying data structure.
* `Stack.peek`: return a copy of the node from the **back** of the underlying data structure.

We can also make a similar argument for `Queue.dequeue` and `Stack.pop`. They appear identical because we chose the front of the underlying data structure as the operating point for the stack.

Finally, we note that `Queue.enqueue` behaves exactly as `LinkedList.add`. Which makes `Stack.push` the only method that we *may* have to write from scratch, because *our choice* of the operating point for the stack is the front of the list. If we added and removed data from the back of the stack, the `push` method would have been identical to the `enqueue` method. For now we leave things as they are, and we just write `push` in the stack class.

This analysis present us with a design dilemma. Do we modify the base class `LinkedList` to include the shared functionality we observed above? Or do we leave `LinkedList` intact, extend it into an intermediate class implementing the shared functionality, and then use the intermediate class to implement the stack and the queue?

For illustrative purposes, we are leaving `LinkedList` intact, even though it is a bit incomplete: it has a method to add nodes but not to remove any of them. We'll take care of this omission in the intermediate class.



## Intermediate class `EnhancedLinkedList`


```text
+-------------------------------+                 +------------+
| EnhancedLinkedList            |  <--- is a -----+ LinkedList |
+-------------------------------+                 +------------+
| + remove(int) -> Node | None  |
| + size() -> int               |  
| + is_empty() -> bool          |
| + inspect(int) -> Node | None |
+-------------------------------+
```

* `EnhancedLinkedList.remove`: removes and returns the i-th node in the linked list, with `i=0` the `head` node and `i=size()-1` the `tail` node.

* `EnhancedLinkedList.inspect`: Returns a copy of the i-th node in the linked list, with `i=0` the `head` node and `i=size()-1` the `tail` node.


## A more realistic approach

Class `LinkedList` should implement the following methods:

* `size() -> int:`: returns the number of nodes in the list.

* `is_empty() -> bool:` evaluates to `True` when `size()==0`.

* `insert(node:Node, i:int)`: inserts a node object at the i-th position of the list. For `i=0` the new node becomes the new head of the list. For `i=size()` the new node is appended at the list.

* `remove(i:int) -> Node | None`: removes and returns the i-th node from the list. The `head` node is at `i=0` and the `tail` node is at `i=size()-1`.

* `inspect(i:int) -> Node | None`: returns a copy of the i-th node from the list. The `head` node is at `i=0` and the `tail` node is at `i=size()-1`.

This eliminates the need for `EnhancedLinkedList`.

In this realistic scenario, base class methods map to ADT methods as follows.

| ADT method<br/> (`Stack` or `Queue`) | Base class call<br/>(`LinkedList`) |
| ---------- | ----------------|
| `Stack.push(node)` | `insert(node, 0)`  |
| `Stack.pop` | `remove(0)`  |
| `Stack.peek` | `inspect(0)`  |
| `Queue.enqueue(node)` | `insert(node, self.size())`  |
| `Queue.dequeue(node)` | `remove(0)`  |
| `Queue.peek` | `inspect(0)`  |
| `*.is_empty` | `is_empty`  |
| `*.size` | `size`  |
