# Asymptotic notation

Given a constant $c>0$ and some natural number $n_0$, a function $T(n)$ is asymptotically upper-bounded by function $f(n)$ if $T(n) \leq cf(n)$ for any $n>n_0$. We write $T(n) \in \mathcal O(f(n))$ and we say that $T(n)$ *is of $\mathcal O(f(n))$*. The formal definition is:

$$
\begin{align*}
\exists\, c>0,\, n_0>0:\, T(n) & \in \mathcal O(f(n)) \Leftrightarrow T(n) \leq cf(n),\, \forall n>n_0
\end{align*}
$$

The context here is as follows:
* $n$ is the size of a problem we wish to solve
* $T(n)$ is the time it takes to solve the problem of size $n$
* $\mathcal O(f(n))$ is the worst case scenario for solving the problem.

For example, if $T(n)=3n+11$ we can show that $T(n)\in\mathcal O(f(n))$ with $f(n)=n$. 

$$
\begin{align*}
T(n) & \in \mathcal O(f(n)) \Rightarrow \\
T(n) & \leq c f(n)  \Rightarrow \\
3n+11 & \leq cn  
\end{align*}
$$
Any $c>3$ would suffice. Consider, for example $c=3.01$. We get

$$
\begin{align*}
3n+11 & \leq 3.01n \Rightarrow \\
11 & \leq 0.01n \Rightarrow \\
n & \geq 1100
\end{align*}
$$
We just showed that for $n_0=1100$ and $c=3.01$, $3n+11\leq cn$ for $n>n_0$. Therefore, $3n+11\in\mathcal O(n)$. 

Another good choice would be $c=4$, yielding $n_0=11$.




In general, any polynomial of the form

$$
T(n) = a_d n^d + a_{d-1} n^{d-1} + a_{d-2} n^{d-2} + \ldots + a_2 n^2 + a_1 n + a_0
$$
is asymptotically upper-bounded by $n^d$, i.e.,

$$T(n)\in\mathcal O(n^d)$$

Such polynomial bounds have three special cases:

* $d=0$ is the class of constant-time problems.
* $d=1$  is the class of linear-time problems.
* $d=2$  is the class of quadratic-time problems.

Broadly speaking there are two major families of upper-bounds for the time required to solve a problem of size $n$:
* $\mathcal O(n^d)$ for problems that require polynomial time; and,
* $\mathcal O(b^n)$ for problems that require exponential time. Here $b>1$.

This distinction is central to the theory of computational complexity. Polynomial problems are considered tractable, and exponential problems intractable. Tractability here means that a known solution can be computed within a practically reasonable time.

There are cases, however, when an exponential problem may be tractable and a polynomial may not. Consinder for example, two problems that can be solved in $\mathcal O(n^{10})$ and $\mathcal O(1.01^n)$ respectively. For a small problem size of $n=2$, the polynomial solution requires 1,024 steps while the expoential solution requires 1.0201 steps. Which one is actually more tractable?


Polynomial and exponential time are not the only two families of upper-bounds. Some problem of size $n$ can be solved in:

* **Sub-linear time** $\mathcal{O}(\log n)$ or  $\mathcal{O}(\sqrt{n})$; problems in this caterogy include binary search, some number-theoretic algorithms.

* **Linearithmic time** $\mathcal{O}(n \log n)$; problems in this caterogy include merge sort, heaps to name a few.

* **Factorial time** $\mathcal{O}(n!)$ which is *worse* than $\mathcal O(2^n)$; problems in this caterogy include permutation problems and brute-force TSP.

* **Quasi-polynomial time** $\mathcal{O}(n^{\log n})$. This family sits at the boundary between polynomial and exponential time. There are few examples for this class. A recent (2016) discovery is graph isomorphism that was shown by Laci Babai to be in $\mathcal{O}(2^{(\log n)^c})$.
