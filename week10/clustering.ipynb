```python
import random
import string

def random_string(min_len: int, max_len: int) -> str:
    length = random.randint(min_len, max_len)
    return ''.join(random.choices(string.ascii_uppercase, k=length))
```

```python

def simple_hash(s: str) -> int:
    return sum(ord(char) for char in s)
```

```python


num_samples = 500000
min_len = 5
max_len = 50

samples = [random_string(min_len, max_len) for _ in range(num_samples)]
hash_values = [simple_hash(s) for s in samples]

plt.figure(figsize=(10, 6))
plt.hist(hash_values, bins=30, edgecolor="black")
plt.title("Distribution of simple_hash Values")
plt.xlabel("Hash value")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.3)
plt.show()
```
